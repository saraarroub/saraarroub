# Partie A — Importation du Dataset, Arrays, Premières Manipulations

### A1. Importation du dataset

### F29. Variance par espèce
Identifiez l’espèce avec la plus grande variance globale et visualisez ses distributions.


```python
# Votre code ici
max_var = -1 # Initialise la variance maximale
winner_species = "" # Initialise le nom de l'espece gagnante
winner_data = None # Initialise les donnees de l'espece gagnante
for name in species_names: # Boucle sur les noms des especes
    data_species = X_clean_id[Y == name] # Selectionne les donnees pour l'espece courante
    global_var = np.sum(np.var(data_species, axis=0)) # Calcule la variance globale
    print(f"- {name} : {global_var:.4f}") # Affiche la variance
    if global_var > max_var: # Verifie si c'est la variance la plus elevee
        max_var = global_var # Met a jour la variance maximale
        winner_species = name # Met a jour l'espece gagnante
        winner_data = data_species # Met a jour les donnees de l'espece gagnante
print(f"\nEspece avec la plus grande variance : {winner_species}") # Affiche l'espece avec la plus grande variance
fig, axs = plt.subplots(2, 2, figsize=(10, 8)) # Cree une grille de 2x2 sous-graphiques
axs_flat = axs.flatten() # Aplatit le tableau d'axes
fig.suptitle(f"Distributions des caracteristiques de : {winner_species}", fontsize=16) # Definit le titre principal
for i in range(4): # Boucle sur les 4 caracteristiques
    axs_flat[i].hist(winner_data[:, i], bins=15, color=colors[3], edgecolor='black', alpha=0.7) # Cree un histogramme
    axs_flat[i].set_title(features[i]) # Definit le titre du sous-graphique
    axs_flat[i].set_ylabel("Frequence") # Etiquette y
plt.show() # Affiche les graphiques
```
